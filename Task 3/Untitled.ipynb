#### Univarite Analysis

1] Outlier Treatment 

```python
def outlier_plot(data):
    cols = data.select_dtypes(include = ['int','float']).columns
    for i,col in enumerate(cols,1):
        plt.subplot(2,2,i)
        sns.boxplot(data[col])
        plt.title(col)
    plt.tight_layout()
    plt.show()
        
```

```python
outlier_plot(raw_data)
```

```python
def outlier_clapping(data):
    cols = data.select_dtypes(include = ['int','float']).columns
    for i in cols:
        q3 = data[i].quantile(0.75)
        q1 = data[i].quantile(0.25)
        
        iqr = q3-q1
        
        upper = q3 + 1.5 * iqr
        lower = q1 - 1.5 * iqr
        
        data[i].clip(lower,upper,inplace = True)
        
```

```python
outlier_clapping(raw_data)
```

```python
outlier_plot(raw_data)
```

### Modelling 

#### 1] Separating Feature and Traget 

```python
feature = raw_data['TV']
target = raw_data['Sales']
```

```python
feature_train,feature_test,target_train,target_test = train_test_split(feature,target,test_size=0.3,random_state=100)
```

```python
feature_train = sm.add_constant(feature_train)
feature_test = sm.add_constant(feature_test)
```

#### 2] Model Building

```python
lr = LinearRegression()
lr.fit(feature_train,target_train)
y_pred = lr.predict(feature_test)
```

#### 3] Model Evaluation

```python
print('The performance evaluation of Logistic Regression is below: ', '\n')
print('Mean squared error: ',mean_squared_error(target_test, y_pred))
print('Mean absolute error: ',mean_absolute_error(target_test, y_pred))
print('R2 score: ',r2_score(target_test, y_pred))
print('\n', '='*100, '\n')

```
